```python
import pandas as pd 
import numpy as np 
from sklearn import linear_model 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('canada_per_capita_income (1).csv')
data
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(data.year,data.PCI,color='purple',marker='o')
```

```python
new_data = data.drop('PCI',axis=1)  
new_data
```

```python
PCI = data.PCI
PCI
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_data,PCI) 
```

```python
reg.get_params()
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg.predict([[2020]])[0]
```
